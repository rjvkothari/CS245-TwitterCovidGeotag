```python
states = [("AL","Alabama"),("AK","Alaska"),("AZ","Arizona"),("AR","Arkansas"),("CA", "California"),("CO", "Colorado"),
("CT","Connecticut"),("DC","Washington DC"),("DE","Delaware"),("FL","Florida"),("GA","Georgia"),
("HI","Hawaii"),("ID","Idaho"),("IL","Illinois"),("IN","Indiana"),("IA","Iowa"),("KS","Kansas"),("KY","Kentucky"),
("LA","Louisiana"),("ME","Maine"),("MD","Maryland"),("MA","Massachusetts"),("MI","Michigan"),("MN","Minnesota"),
("MS","Mississippi"),("MO","Missouri"),("MT","Montana"),("NE","Nebraska"),("NV","Nevada"),("NH","New Hampshire"),
("NJ","New Jersey"),("NM","New Mexico"),("NY","New York"),("NC","North Carolina"),("ND","North Dakota"),("OH","Ohio"),
("OK","Oklahoma"),("OR","Oregon"),("PA","Pennsylvania"),("RI","Rhode Island"),("SC","South Carolina"),("SD","South Dakota"),
("TN","Tennessee"),("TX","Texas"),("UT","Utah"),("VT","Vermont"),("VA","Virginia"),("WA","Washington"),("WV","West Virginia"),
("WI","Wisconsin"),("WY","Wyoming")]
statelist = set([j for i,j in states ])
ablist = set([i for i,j in states ])
d = {}
for i,j in states:
    d[i] = j
```

```python
import pandas as pd
import glob

path = r'user-tweets' # use your path
all_files = glob.glob(path + "/*.csv")

li = []
s = 0
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    s+=len(df)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
print(s)
```

```python
len(frame)
```

```python
frame = frame[~frame['location'].isin(['USA','United States','United States of America'])]
#frame['state'] = frame['location'].apply(lambda x : x.split(",")[0] if len(x.split(",")) == 2 else x)
```

```python
import re
frame['inputlocation'] = frame['location']
frame['location'] = frame['location'].apply(lambda x : x.replace('USA',''))
frame['location'] = frame['location'].apply(lambda x : x.replace('United States',''))
frame['location'] = frame['location'].apply(lambda x : x.replace('United States of America',''))

#frame['location'] = frame['location'].apply(lambda x : x.replace(" ",""))
```

```python
len(frame)
```

```python
frame['Newstate'] = ''
frame['splitcount'] = 0
for index, row in frame.iterrows():
    if row['location']:
        row['location'] = re.sub(r'\s+$','',row['location'])
        row['location'] = re.sub(r',$','',row['location'])
        #row['location'] = re.sub(r',\s+','',row['location'])
        #row['location'] = re.sub(r'\s+,','',row['location'])
        #row['location'] = re.sub(r'\s*','',row['location'])
    if row['location'] and  row['location'][-1] == ',':
        row['location'] = row['location'][:-1]
    split = row['location'].split(",")
    for i in range(len(split)):
        if split[i]:
            split[i] = split[i].strip()
    
    frame['splitcount'][index] = len(split)
    if len(split) == 1 :
        frame['Newstate'][index] = row['location']
        
    elif len(split) == 2 :
        if split[0] in statelist or split[0] in ablist:
            frame['Newstate'][index]  = split[0]
        elif split[1] in statelist or split[1] in ablist:
            frame['Newstate'][index]  = split[1]
            
    elif len(split) == 3 :
        if split[0] in statelist or split[0] in ablist:
            frame['Newstate'][index]  = split[0]
        elif split[1] in statelist or split[1] in ablist:
            frame['Newstate'][index]  = split[1]
        elif split[2] in statelist or split[2] in ablist:
            frame['Newstate'][index]  = split[3]
                    
    elif len(split) == 4 :
        if split[0] in statelist or split[0] in ablist:
            frame['Newstate'][index]  = split[0]
        elif split[1] in statelist or split[1] in ablist:
            frame['Newstate'][index]  = split[1]
        elif split[2] in statelist or split[2] in ablist:
            frame['Newstate'][index]  = split[2]
        elif split[3] in statelist or split[3] in ablist:
            frame['Newstate'][index]  = split[3]

    elif len(split) == 5 :
        if split[0] in statelist or split[0] in ablist:
            frame['Newstate'][index]  = split[0]
        elif split[1] in statelist or split[1] in ablist:
            frame['Newstate'][index]  = split[1]
        elif split[2] in statelist or split[2] in ablist:
            frame['Newstate'][index]  = split[2]
        elif split[3] in statelist or split[3] in ablist:
            frame['Newstate'][index]  = split[3]
        elif split[4] in statelist or split[4] in ablist:
            frame['Newstate'][index]  = split[4]
        
```
